```python
#Data Cleaning  (Replace '--'  to NaN)
import numpy as np
df.replace("--", np.nan,regex = True, inplace = True)
```

```python
#Check Null Values In The Dataset
df.isnull().sum()
```

```python
sns.heatmap(df.isnull())
```

```python
df.dropna(axis=0, inplace = True)
```

```python
sns.heatmap(df.isnull())
```

```python
#Data Cleaning [ Rank Column ]
df["Rank"].head()
```

```python
df["Rank"] = df["Rank"].str.replace("st", "")
df["Rank"] = df["Rank"].str.replace("nd", "")
df["Rank"] = df["Rank"].str.replace("th", "")
df["Rank"] = df["Rank"].str.replace("rd", "")
df["Rank"] = df["Rank"].str.replace(",", "")
```

```python
df.head()
```

```python
df["Rank"] = df["Rank"].astype(int)
```

```python
df.tail()
```

```python
#Data Cleaning [ Video Uploads & Subscribers ]
df["Video Uploads"].dtype
```

```python
df["Subscribers"].dtype
```

```python
df["Video Uploads"] = df["Video Uploads"].astype(int)
```

```python
df["Subscribers"] = df["Subscribers"].astype(int)
```

```python
df["Video Uploads"].dtype
```

```python
df["Subscribers"].dtype
```

```python
#Data Cleaning [ Grade Column ]
df["Grade"].unique()
```

```python
df["Grade"] = df["Grade"].map({"A++ ":5, "A+ ":4, "A ":3, "A- ":2, "B+ ":1})
```

```python
df.head(2)
```

```python
#Find Average Views For Each Channel
df["Video views"]/df["Video Uploads"]
```

```python
#Find Out Top Five Channels With Maximum Number of Video Uploads
x = df.sort_values(by = "Video Uploads", ascending = False).head(5)
x
```

```python
x["Channel name"]
```

```python
# Find Correlation Matrix
df.corr()
```

```python
#Which Grade Has A Maximum Number of Video Uploads?
n = df["Video Uploads"].max()
df[df["Video Uploads"] == n]["Grade"]
```

```python
#Which Grade Has The Highest Average Views?
df.groupby("Grade")["Video views"].mean().sort_values(ascending = False)
```

```python
#Which Grade Has The Highest Number of Subscribers?
n = df["Subscribers"].max()
df[df["Subscribers"] == n]["Grade"]
```

```python
#Which Grade Has The Highest Video Views?
n = df["Video views"].max()
df[df["Video views"] == n]["Grade"]
```
